# Mapa Conceptual 

Aquí tienes el código de Python utilizando la librería `graphviz` para generar el mapa conceptual de la secuencia de tu modelo predictivo para el dengue.

He organizado el flujo utilizando `subgraph` (clústeres) para que cada etapa del pipeline quede visualmente agrupada y ordenada secuencialmente.



```python
from graphviz import Digraph

# Crear el objeto Digraph
dot = Digraph('Modelo_Predictivo_Dengue', comment='Secuencia de Constitución de Modelo Predictivo')

# Configuración global del gráfico
dot.attr(rankdir='TB', size='10,12', compound='true')
dot.attr('node', shape='box', style='filled, rounded', color='#2B3E50', fontname='Helvetica', fontcolor='white', fillcolor='#34495E')
dot.attr('edge', fontname='Helvetica', fontsize='10', color='#7F8C8D')

# --- ETAPA 1: Obtención de Datos ---
with dot.subgraph(name='cluster_0') as c:
    c.attr(label='1. Obtención de Datos', bgcolor='#EAECEE', style='filled', color='#BDC3C7')
    c.node('pandas', 'Lectura y Estructuración\n(pandas)', fillcolor='#1F618D')

# --- ETAPA 2: Reducción Dimensional ---
with dot.subgraph(name='cluster_1') as c:
    c.attr(label='2. Reducción Dimensional / Selección de Atributos', bgcolor='#EAF2F8', style='filled', color='#A9CCE3')
    c.node('pca', 'PCA\n(Lineal/No supervisado)', fillcolor='#2980B9')
    c.node('pearson', 'Correlación de Pearson\n(Lineal)', fillcolor='#2980B9')
    c.node('spearman', 'Correlación de Spearman\n(No lineal/Monótona)', fillcolor='#2980B9')
    c.node('lasso', 'Regularización Lasso\n(L1)', fillcolor='#2980B9')
    c.node('elastic', 'Elastic Net\n(L1 + L2)', fillcolor='#2980B9')

# --- ETAPA 3: Escalamiento de Atributos ---
with dot.subgraph(name='cluster_2') as c:
    c.attr(label='3. Escalamiento de Atributos', bgcolor='#E8F8F5', style='filled', color='#A3E4D7')
    c.node('scaler', 'StandardScaler\n(sklearn)', fillcolor='#16A085')

# --- ETAPA 4: Entrenamiento del Modelo ---
with dot.subgraph(name='cluster_3') as c:
    c.attr(label='4. Entrenamiento del Modelo (Regresión)', bgcolor='#FEF9E7', style='filled', color='#F9E79F')
    c.node('dt', 'Árbol de Decisión', fillcolor='#D4AC0D')
    c.node('rf', 'Random Forest', fillcolor='#D4AC0D')
    c.node('xgboost', 'XGBoost / Gradient Boosting', fillcolor='#D4AC0D')
    c.node('lgbm', 'LightGBM', fillcolor='#D4AC0D')
    c.node('svm', 'SVM (SVR)', fillcolor='#D4AC0D')
    c.node('nn', 'Redes Neuronales\n(MLP / Deep Learning)', fillcolor='#D4AC0D')

# --- ETAPA 5: Medición de Desempeño ---
with dot.subgraph(name='cluster_4') as c:
    c.attr(label='5. Medición de Desempeño', bgcolor='#FADBD8', style='filled', color='#F5B7B1')
    c.node('mae', 'MAE\n(Error Absoluto Medio)', fillcolor='#C0392B')
    c.node('rmse', 'RMSE\n(Raíz del Error Cuadrático Medio)', fillcolor='#C0392B')

# --- CONEXIONES (Flujo del pipeline) ---

# De Obtención de Datos a las opciones de Reducción Dimensional
reducciones = ['pca', 'pearson', 'spearman', 'lasso', 'elastic']
for r in reducciones:
    dot.edge('pandas', r)

# De Reducción Dimensional a Escalamiento
for r in reducciones:
    dot.edge(r, 'scaler')

# De Escalamiento a las opciones de Algoritmos
algoritmos = ['dt', 'rf', 'xgboost', 'lgbm', 'svm', 'nn']
for alg in algoritmos:
    dot.edge('scaler', alg)

# De Algoritmos a las Métricas de Desempeño
metricas = ['mae', 'rmse']
for alg in algoritmos:
    for m in metricas:
        dot.edge(alg, m)

# Para visualizarlo en un entorno Jupyter o guardar el archivo:
# dot.render('pipeline_dengue', format='png', cleanup=True)
dot.source

dot.render(r"C:\Users\marco\Documentos\investigacion\machine_learning_idalina\0_plan_de_clubes_de_revista\mapa_estrutura_constitucion_modelo.png",\
    format = 'png')
```

```python

### Características del diseño:

* **Estructura en Bloques (Clústeres):** Cada etapa del pipeline de Machine Learning tiene su propio contenedor con un fondo de color sutil para diferenciar las fases del proyecto.
* **Flujo Secuencial:** El gráfico se lee de arriba a abajo (`rankdir='TB'`), ramificándose en las opciones disponibles y volviéndose a unificar en los pasos obligatorios (`pandas` -> *Opciones de reducción* -> `StandardScaler` -> *Opciones de modelos* -> *Métricas*).
* **Código limpio:** Utiliza bucles en Python para conectar eficientemente los nodos multiplexados, evitando declarar líneas repetitivas una por una.
```
