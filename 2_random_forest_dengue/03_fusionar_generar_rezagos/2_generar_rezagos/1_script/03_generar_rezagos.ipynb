```python
# Inicio del análisis
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np  
```

# Adicionar al dataset los atributos meteorológicos rezagados con un rezago hasta de 12 semanas. 

```python
ubicacion_datos = r"C:\Users\marco\Documentos\investigacion\arima\2_ejecucion_plan_trabajo\03_fusionar_generar_rezagos\2_generar_rezagos\2_datos\1_raw\datos_semanal_meteo_completos_epi_sin_rezagos.xlsx"
df_semanal_meteo_epi= pd.read_excel(ubicacion_datos)
df_semanal_meteo_epi.head()
```

```python
# Lista de variables meteorológicas
variables_locales = ['temp', 'temp_max', 'temp_min',
       'hum_esp', 'hum_rel', 'prec', 'dias_lluvia', 'vel_vi', 'vel_vi_max',
       'vel_vi_min', 'uv'
]

variables_macroclimaticas = ['soi', 'sst']

# Número de rezagos para las variables locales y macroclimáticas
max_lag_local = 8

# Definimos el inicio y fin de los rezagos macroclimáticos (de 8 a 12)
min_lag_macro = 8
max_lag_macro = 12

# Iniciar el DataFrame de rezagos a partir del original
df_semanal_meteo_epi_rezagos = df_semanal_meteo_epi.copy()

# Generar rezagos para variables locales (del lag 1 al 8)
for var in variables_locales:
    for lag in range(1, max_lag_local + 1):
        df_semanal_meteo_epi_rezagos[f"{var}_lag_{lag}"] = df_semanal_meteo_epi_rezagos[var].shift(lag)

# Generar rezagos para variables macroclimáticas (del lag 8 al 12)
for var in variables_macroclimaticas:
    for lag in range(min_lag_macro, max_lag_macro + 1):
        df_semanal_meteo_epi_rezagos[f"{var}_lag_{lag}"] = df_semanal_meteo_epi_rezagos[var].shift(lag)

# Ver resultado
df_semanal_meteo_epi_rezagos.head(15)
```

```python
df_semanal_meteo_epi_rezagos.columns    
df_semanal_meteo_epi_rezagos.tail(5)
```

```python
# Eliminar filas con valores NaN (debido a los rezagos)
df_semanal_meteo_epi_rezagos = df_semanal_meteo_epi_rezagos.dropna().reset_index(drop=True) 
df_semanal_meteo_epi_rezagos.head(5)
```

```python
# Guardar el DataFrame con rezagos
df_semanal_meteo_epi_rezagos = df_semanal_meteo_epi_rezagos.dropna().reset_index(drop=True) 
df_semanal_meteo_epi_rezagos.to_excel(r"C:\Users\usuario1\OneDrive - Universidad de Antioquia\UNIVERSIDAD DE ANTIOQUIA\Proyecto SAT Dengue\Bases de datos\datos_consolidados\1_datos_semana_epi_meteo\datos_semanal_meteo_epi_con_rezagos\datos_semanal_meteo_epi_con_rezagos.xlsx", index=False)
```
